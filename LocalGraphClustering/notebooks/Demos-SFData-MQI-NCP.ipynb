```python
import localgraphclustering as lgc
import numpy as np
import matplotlib.pyplot as plt
```

```python
g=lgc.GraphLocal()
g.read_graph(
       "datasets/senate.edgelist",
    separator= " ")
xy = np.loadtxt("datasets/senate.xy")
g._num_vertices
```

```python
p = g.draw(xy, figsize=(10,4),nodecolor='white',alpha=0.2)
```

```python
ncp = lgc.NCPData(g).mqi(nthreads=2,timeout=60) 
```

```python
p = lgc.NCPPlots(ncp)
p.mqi_input_output_cond_plot();
```
